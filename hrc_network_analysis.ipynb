```python
# new datframe (hf) that excludes email with the MetaDataFrom value missing
hf = df[df['MetadataFrom'] != '']
hf
```

```python
# in the hf dataframe only show the columns in features and rename MetadataTo to Alias to merge
# with the Aliases df

features = ['MetadataTo','MetadataFrom','SenderPersonId','ExtractedBodyText']

hf = hf[features]
hf = pd.DataFrame(hf)
hf
#unique_names_hf = hf.MetadataTo.unique()
#len(unique_names_hf)
#unique_names_hf
```

```python
# Create a new dataframe from hf that contains the name (MetadataFrom) and the PersonID
# Add only unique values in this dataframe
names_df = hf
names_features = ['MetadataFrom','SenderPersonId']
names_df = names_df[names_features]
names_df = names_df.drop_duplicates()
names_df
```

```python
# In names_df, change MetadataFrom column name to MetaDataTo, so as to merge with hf

names_df = names_df.rename(columns =({'MetadataFrom':'MetadataTo', 'SenderPersonId':'ReceiverPersonId'}))
names_df

#len(names_df)
#289 names
```

```python
# Merge hf dataframe with names_df on MetadataTo column

merged_hf = hf.merge(names_df, on='MetadataTo', how='left')
merged_features = ['MetadataTo','MetadataFrom','ReceiverPersonId','SenderPersonId','ExtractedBodyText']
merged_hf = merged_hf[merged_features]
merged_hf
```

```python
# The merged_hf dataframe is missing many 1,641 emails (174 names). They are identified below

find_nulls = merged_hf[merged_hf.ReceiverPersonId.isnull()]
#find_nulls.drop_duplicates('MetadataTo')
find_nulls_feature = ['MetadataTo']
find_nulls = find_nulls[find_nulls_feature]
find_nulls = find_nulls.drop_duplicates()
find_nulls
#174 names missing/1,641 emails
```

```python
# Merge the find_nulls dataframe with the Alias table to recover more PersonIds
#Load aliases dataframe
aliases_df = pd.read_csv('data_files/Aliases.csv')

#Only include Alias and PersonId columns from Aliases df
aliases_df
aliases = ['Alias','PersonId']
aliases = aliases_df[aliases]

# Rename Alias column to be MetadataTo
aliases = aliases.rename(columns =({'Alias':'MetadataTo', 'PersonId':'ReceiverPersonId'}))
aliases
```

```python
# Merge the find_nulls dataframe with the Alias dataframe to get PersonId
# 95 emails still without a PersonId

fix_nulls_df = find_nulls.merge(aliases, on='MetadataTo', how='left')
fix_nulls_df = fix_nulls_df[fix_nulls_df.ReceiverPersonId.notnull()]
fix_nulls_df
#len(fix_nulls_df)
#79 names
```

```python
# Append fix_nulls_df to names_df to have a comprehensive list of names and PersonIds
top_stack = names_df
top_stack
bottom_stack = fix_nulls_df#.reset_index()
bottom_stack
concat_names_df = pd.concat([top_stack, bottom_stack], axis=0)
concat_names_df[concat_names_df.ReceiverPersonId.notnull()]

concat_names_df = concat_names_df.rename(columns =({'ReceiverPersonId':'ReceiverPersonIdConct'}))
concat_names_df
#len(concat_names_df)
#368 names
```

```python
# Merge the merged_hf table with the concat_names_df table to fix for missing values

merged_hf = merged_hf.merge(concat_names_df, on='MetadataTo', how='left')
merged_hf

```

```python
hrc_dict={}

for tup in list(to_from_df.Sender_Receiver.values):
    #print tup #to_from_dict
    if tup not in hrc_dict:
        hrc_dict[tup] = 1
    else:
        hrc_dict[tup] = hrc_dict[tup] + 1   

print hrc_dict
```

```python
# Order hrc_dict to find who sent the most emails and to whom
from collections import OrderedDict
od = OrderedDict(sorted(hrc_dict.items(), key=lambda x:x[1], reverse=True))
od
```

```python
plt.bar(range(len(hrc_dict)), hrc_dict.values(), align='center')
plt.xticks(range(len(hrc_dict)), hrc_dict.keys())

plt.show()
```

```python
hrc_graph_dict = {}
for s, r in hrc_dict:
    if hrc_dict[s,r] <= 10:
        pass
    else:
        hrc_graph_dict[s,r]=hrc_dict[s,r]

hrc_graph_dict
        
    
```

```python
G = nx.Graph()

# To add the a node for each of the email parties:
# srcs, dests = zip(* [(fr, to) for (fr, to) in hrc_graph_dict.keys()])
# G.add_nodes_from(srcs + dests)
# (but it isn't needed IF the following method is used
#  to add the edges, since add_edge also creates the nodes if
#  they don't yet exist)

# note the use of the attribute "weight" here
for (s,r), count in hrc_graph_dict.items():
    G.add_edge(s, r, weight=count)

# produce info to draw:
# a) if weight was used above, spring_layout takes 
#    into account the edge strengths
pos = nx.spring_layout(G)

# b) specifiy edge labels explicitly
# method from https://groups.google.com/forum/#!topic/networkx-discuss/hw3OVBF8orc
edge_labels=dict([((u,v,),d['weight'])
for u,v,d in G.edges(data=True)])

# draw it
plt.figure(1);
#plt.figure(num=None, figsize=(10, 10), dpi=1200);
nx.draw_networkx(G, pos, with_labels=True)
nx.draw_networkx_edge_labels(G,pos,edge_labels=edge_labels)
nx.draw(G, node_size=.01)

plt.axis('equal') # spring weighting makes more sense this way
plt.show()
```
